### Mengimport mengimport semua library yang mau dipakai, lalu memanggil csv dan menampilkan

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

alpi = pd.read_csv('music_genre.csv')
alpi.head()
```

### Menjadikan obtained_date sebagai index

```python
alpi.set_index('obtained_date', inplace=True)
alpi.head()
```

```python
alpi.loc['4-Apr','music_genre']
```

### membuat variabel "status_alpi" dan "alpi_group' agar enak pemanggilannya

```python
status_alpi = alpi['music_genre'].value_counts()
status_alpi
```

```python
alpi_group = alpi.groupby('music_genre').sum()
alpi_group.head()
```

### dalam grafik batang, untuk disemua jenis genre totalnya semua sama diangka 5000

```python
status_alpi.plot.bar()
plt.title('Genre musik')
plt.xlabel('jenis')
plt.ylabel('angka genre')
plt.show()
```

### dalam histogram, popularitas paling tinggi ada diangka 12.000

```python
plt.hist(alpi['popularity'])
plt.title('popularitas')
plt.xlabel('jenis')
plt.ylabel('banyak popularitas')
plt.show()
```

### dalam pie chart, persenan yang paling tinggi jatuh ke genre rock.

```python
plt.figure(figsize=(7,12))
alpi_group['popularity'].plot.pie(autopct='%.1f%%')
plt.title('persenan popularitas')
```

### terlihat di heatmap, korelasi yang mendominasi atau mendekati angka 1 adalah popularity 

```python
# seaborn
alpi_corr = alpi.loc[:, 'instance_id':'energy'].corr()
alpi_corr 
```

```python
sns.heatmap(alpi_corr , annot=True)
```
